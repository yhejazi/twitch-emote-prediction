```python
import pandas as pd
import re
import numpy as np
import os
import warnings

warnings.filterwarnings('ignore')
```

```python
top_emotes = ["Kappa", "LUL", "TriHard", "PogChamp", "MonkaS", "VoHiYo", "KEKW", "PepeLaugh", "4Head", 
              "CmonBruh", "Booba", "<3", "haHaa", "SourPLS", "MonkaW", "PauseChamp", "4Weird", "Sadge", 
              "gachiGASM", "Pepehands", "Jebaited", "COGGERS", "MingLee", "WeirdChamp", "MonkaGIGA", 
              "GayPride", "PJSalt", "MonkaH", "BlessRNG", "NotLikeThis"]


more_emotes = ["ForsenCD", "5Head", "BibleThump", "Pog", "SMOrc", "MrDestructoid", "MonkaHmm", "FailFish",
              "HeyGuys", "SquadW", "gachiBASS", "PRChase", "CurseLit", "FeelsBadMan", "FeelsGoodMan", "Kreygasm",
              "SwiftRage", "Bloodtrail", "PogU", "TheIlluminati", "Poggers", "DansGame", "duDudu", "BabyRage",
              "SabaPing", "MorphinTime", "ResidentSleeper", "Seemsgood", "monkaOMEGA", "Wutface", "AYAYA", 
              "TheThing", "SingsMic", "Keepo", "SSSsss", "BrokeBack", "widepeepoHappy", "CoolStoryBob", "3Head",
              "YEP", "Pepega", "Kkona", "KKomrade", "FeelsBirthdayMan", "DoritosChip"]

top_emotes.extend(more_emotes)
```

```python
df = pd.DataFrame()
directory = 'data/raw'

raw_counter = 0
one_emote_counter = 0
 
# iterate over files in the directory
for filename in os.listdir(directory):
    f = os.path.join(directory, filename)
    # checking if it is a file
    if os.path.isfile(f):
        data = pd.read_csv(f) 
        print('File:', filename, '- rows:', len(data))

        # Count each emote into their own columns. Total unique emotes
        data['total'] = 0
        data['unique_emotes'] = 0
        data['label'] = 'Kappa' 
        data['input'] = data['message']
        for emote in top_emotes:
            reg = r'\b' + emote + r'\b'
            data[emote] = data['message'].str.count(reg)
            data[emote + '_exists'] = np.sign(data[emote])
            data['total'] = data['total'] + data[emote]
            data['unique_emotes'] = data['unique_emotes'] + data[emote + '_exists']
            # Update label 
            data.loc[data[emote] > 0, 'label'] = emote
            # Remove emote from input
            data['input'] = data['input'].replace(reg, r'', regex=True)

        raw_counter = raw_counter + len(data)
        ### Cleanup ###
        # Filter out messages that 
        # - contain more than one kind of emote
        # - do not contain an emote
        data = data[(data.unique_emotes <=1) & (data.total > 0)]
        print('1 unique emote:', len(data))
        one_emote_counter = one_emote_counter + len(data)

        # Keep messages at least tokens long
        data['message_length'] = data['input'].str.split().str.len()
        data = data[data.message_length >= 3]
        print('Length >= 3:', len(data))
        df = pd.concat([df, data])
    
```

```python
print('Number of raw rows scraped:', raw_counter)
print('Number of rows with one unique emote:', one_emote_counter)
print('Number of final rows:', len(df))
```
